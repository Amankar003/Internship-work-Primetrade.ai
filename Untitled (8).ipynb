```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from scipy.stats import ttest_ind
```

Load Fear-Greed Data

```python
fg_df = pd.read_csv("fear_greed_index.csv")
```

```python
fg_df
```

```python
fg_df['date'] = pd.to_datetime(fg_df['date'])
fg_df.rename(columns={'Classification': 'Sentiment'}, inplace=True)
```

```python
fg_df
```

Load Trader Data

```python
trader_df = pd.read_csv("historical_data.csv")
```

```python
trader_df
```

```python
trader_df['Timestamp IST'] = pd.to_datetime(trader_df['Timestamp IST'], dayfirst=True)
trader_df['date'] = trader_df['Timestamp IST'].dt.date
trader_df['time'] = trader_df['Timestamp IST'].dt.time
```

```python
trader_df
```

Merge the Data

```python
merged_df = pd.merge(trader_df, fg_df, how='left', left_on='date', right_on='date')
merged_df.dropna(subset=['classification'], inplace= True)
```

```python
merged_df
```

# Basic EDA

### 1. Sentiment Distribution

```python
sns.countplot(data=merged_df, x='classification')
plt.title("Market Sentiment Distribution")
plt.show()
```

### 2. PnL Distribution under sentiment

```python
sns.boxplot(data=merged_df, x='classification', y='Closed PnL')
plt.title("PnL vs Market Sentiment")
plt.show
```

### 3. Average PnL per Sentiment 

```python
print(merged_df.groupby('classification')['Closed PnL'].mean())
```

## Statistical Test

```python
fear_pnl = merged_df[merged_df['classification'] == 'Fear']['Closed PnL']
greed_pnl = merged_df[merged_df['classification'] == 'Greed']['Closed PnL']

t_stat, p_val = ttest_ind(fear_pnl, greed_pnl, equal_var=False)
print(f"T-Statistic: {t_stat}, P-value: {p_val}")
```

## Average Tradesize by Sentiment

```python
avg_trade_size = merged_df.groupby('classification')['Size USD'].mean()
print("Average Trade Size by Market Sentiment:\n", avg_trade_size)
```

```python
avg_trade_size.plot(kind='bar', color=['red', 'orange', 'green', 'blue'])
plt.title("Average Trade Size by Market Sentiment")
plt.ylabel("Average Size (USD)")
plt.xlabel("Sentiment")
plt.xticks(rotation=45)
plt.show()
```
